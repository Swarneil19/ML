```python
dataset = pd.read_csv('heart-disease.csv')
```

```python
dataset.head()
```

```python
dataset['age'].plot.hist(bins = 50);
```

```python
dataset.plot.hist(figsize = (10,40), subplots = True);
```

```python
dataset['chol'].plot.box();
```

```python
dataset
```

```python
over_50 = dataset[dataset['age'] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = 'scatter', 
            x = 'age',
            y = 'chol', 
            c = 'thal');
```

```python
# OO method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter', 
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax );
ax.set_xlim([45,100])
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

#Plotting the data

scatter = ax.scatter(x = over_50['age'],
          y = over_50['chol'],
          c = over_50['target']);

ax.set(title = 'Heart Disease and Cholestrol levels');

ax.legend(*scatter.legend_elements(), title = 'target');

# Add a horizontal line 

ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
fig, ax = plt.subplots(figsize = (10,7))

scatter = ax.scatter(x = over_50['chol'],
           y = over_50['thalach'], 
           c = over_50['target']);

ax.set(title = 'Cholestrol and Max Heart Rate');

ax.legend(*scatter.legend_elements(), title = 'target');

ax.axhline(over_50['thalach'].mean(),
          linestyle = '--');
```

```python
# Subplots of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                             ncols = 1,
                             figsize = (10,10))
# Add data axis 0
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target']);

ax0.set(title = 'Heart disease and cholestrol level',
       xlabel = 'Age',
       ylabel = 'Cholestrol');

# Adding a legend
ax0.legend(*scatter.legend_elements(), title = 'target');

# Add a meanline
ax0.axhline(over_50['chol'].mean(),
           linestyle = '--');

# Add data axis 1
scatter = ax1.scatter(x = over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target']);

ax1.set(title = 'Heart disease and max heart rate',
       xlabel = 'Age',
       ylabel = 'Max Heart Rate');

# Adding a legend
ax1.legend(*scatter.legend_elements(), title = 'target');

#Adding meanline 
ax1.axhline(over_50['thalach'].mean(),
           linestyle = '--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10))

ax0.boxplot(over_50['thalach'])
ax1.hist(over_50['chol'])
```
